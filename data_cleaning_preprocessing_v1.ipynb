```python
#merge covid levels by county dataset and vaccination by county
#want to merge on a combo Date-Location index: will be on a date-state index (vaccination df doesnt have county info)

#need the "Location" column in vaccination_cdc_df to be full state name, not abbreviations
#dictionary mapping state abbreviations to full names
state_abb_dict = {
    'AL': 'Alabama', 'AK': 'Alaska', 'AZ': 'Arizona', 'AR': 'Arkansas', 'CA': 'California',
    'CO': 'Colorado', 'CT': 'Connecticut', 'DE': 'Delaware', 'FL': 'Florida', 'GA': 'Georgia',
    'HI': 'Hawaii', 'ID': 'Idaho', 'IL': 'Illinois', 'IN': 'Indiana', 'IA': 'Iowa',
    'KS': 'Kansas', 'KY': 'Kentucky', 'LA': 'Louisiana', 'ME': 'Maine', 'MD': 'Maryland',
    'MA': 'Massachusetts', 'MI': 'Michigan', 'MN': 'Minnesota', 'MS': 'Mississippi', 'MO': 'Missouri',
    'MT': 'Montana', 'NE': 'Nebraska', 'NV': 'Nevada', 'NH': 'New Hampshire', 'NJ': 'New Jersey',
    'NM': 'New Mexico', 'NY': 'New York', 'NC': 'North Carolina', 'ND': 'North Dakota', 'OH': 'Ohio',
    'OK': 'Oklahoma', 'OR': 'Oregon', 'PA': 'Pennsylvania', 'RI': 'Rhode Island', 'SC': 'South Carolina',
    'SD': 'South Dakota', 'TN': 'Tennessee', 'TX': 'Texas', 'UT': 'Utah', 'VT': 'Vermont',
    'VA': 'Virginia', 'WA': 'Washington', 'WV': 'West Virginia', 'WI': 'Wisconsin', 'WY': 'Wyoming'
}
#create new column 'state' with full state names
vaccination_cdc_df['State'] = vaccination_cdc_df['Location'].map(state_abb_dict)

#need the format of the "Date" column in vaccination_cdc_df to match the format of "date_updated" in community_levels_df
#convert the 'Date' column to datetime objects
vaccination_cdc_df['Date'] = pd.to_datetime(vaccination_cdc_df['Date'], format='%m/%d/%Y')
#format the vaccination_cdc_df dates as strings in the the same format as in community_levels_df
vaccination_cdc_df['Date'] = vaccination_cdc_df['Date'].dt.strftime('%Y-%m-%d')

#want to make compound-index of state and date
#create new column 'date_state_index' by concatenating the 'Date' and 'State' columns with '-' in between
#this new column can be used to uniquely index the vaccindation_cdc_df dataframe
vaccination_cdc_df['date_state_index'] = vaccination_cdc_df['Date'] + '-' + vaccination_cdc_df['State']
vaccination_cdc_df.head(3)
```

```python
#double check that the 'date_state_index' column in vaccination_cdc_df has only unique values
index_counts = vaccination_cdc_df.date_state_index.value_counts()

has_duplicates = False

for count in index_counts:
    if count > 1:
        has_duplicates = True

print(has_duplicates)
```

```python
#double check that vaccination_cdc_df has only 50 states
len(vaccination_cdc_df.State.value_counts())
```

```python
community_levels_df.shape
```

```python
#now want to put the same type of index in community_levels_df that we put into vaccination_cdc_df
#(date_state_index)

#theres a problem -- community_levels_df has all the provinces and territories in addition to states
#vaccination_cdc_df has just the 50 states
#want to drop all rows from community_levels_df where state is not one of the 50 US states

#get a list of the 50 US states
states_list = list(vaccination_cdc_df['State'].unique())

#drop rows of community_levels_df where the state is not one of the 50 US states
community_levels_df = community_levels_df[community_levels_df['state'].isin(states_list)]
#print shape of df to ensure that rows were dropped
community_levels_df.shape
```

```python
#back to putting the same type of index into community_levels_df that we put into vaccination_cdc_df
#want a compound-index of state and date of this format: 2023- 05 - 10-Nebraska

#create new column 'date_state_index' by concatenating the 'Date' and 'State' columns with '-' in between
#this new column can be used to uniquely index the  dataframe community_levels_df
community_levels_df['date_state_index'] = community_levels_df['date_updated'] + '-' + community_levels_df['state']
community_levels_df.head(3)
```

```python
#this date_state_index WILL have duplicates -- the location is being measured on a smaller scale (county)
#therefore there could be multiple observations in a given state on a given day

#thats fine, though, we can still merge in vaccination data for a given state on a given day from vaccination_cdc_df

index_counts = community_levels_df.date_state_index.value_counts()

has_duplicates = False
for count in index_counts:
    if count > 1:
        has_duplicates = True

print(has_duplicates)

```

```python
#adding prefix to the column names of both dataframes: helps distinguish where a feature came from once dfs are merged
community_levels_df = community_levels_df.add_prefix('comm_')
vaccination_cdc_df = vaccination_cdc_df.add_prefix('vacc_')

#dont add prefix to the index that we want to merge on, tho
community_levels_df = community_levels_df.rename(columns={'comm_date_state_index': 'date_state_index'})
vaccination_cdc_df = vaccination_cdc_df.rename(columns={'vacc_date_state_index': 'date_state_index'})

vaccination_cdc_df.head(3)
```

```python
#find common dates between community_levels_df and vaccination_cdc_df
overlap_dates = community_levels_df[community_levels_df['comm_date_updated'].isin(vaccination_cdc_df['vacc_Date'])]
len(overlap_dates['comm_date_updated'].tolist())
```

```python
#find common date_state_index between community_levels_df and vaccination_cdc_df
overlap_dates = community_levels_df[community_levels_df['date_state_index'].isin(vaccination_cdc_df['date_state_index'])]
len(overlap_dates['date_state_index'].tolist())
```

```python
#now merge community_levels_df and vaccination_cdc_df

#merging the dataframes on 'date_state_index'
merged_df = pd.merge(community_levels_df, vaccination_cdc_df, on='date_state_index', how='inner')
merged_df.head(3)
```

```python
merged_df.shape
```

```python
#check how many null values each column in the merged df has
for column in merged_df.columns:
    nan_count = merged_df[column].isna().sum()
    print(f"Number of NaN values in {column}: {nan_count}")
```

```python
#lots of those columns have most or all null values -- drop column with over 1000 null values
#counting null values in each column
nan_counts = merged_df.isna().sum()
#identify columns with more than 1000 NaN values
columns_to_drop = nan_counts[nan_counts > 1000].index.tolist()
#dropping these columns from merged_df
merged_df.drop(columns=columns_to_drop, inplace=True)
print(len(columns_to_drop))
```
