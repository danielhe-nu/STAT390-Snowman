```python
df1 = pd.read_csv('CA.csv', index_col=0)
df1 = df1.drop(columns=['STATE'])
df1 = df1.reset_index()
df1.rename(columns={'PEOPLE_POSITIVE_NEW_CASES_COUNT': 'y'}, inplace=True)
df1.rename(columns={'Date': 'ds'}, inplace=True)
df1 = df1.sort_values(by='ds')
df1
```

```python
train = df1.iloc[:632]
test = df1.iloc[632:]
```

```python
def train_prophet(features, train_data, test_data):
    # Initialize Prophet model
    model = Prophet()
    
    # Add each feature in the current set as a regressor
    for feature in features:
        model.add_regressor(feature, standardize=False)
    
    # Fit the model on the training data
    model.fit(train_data)
    
    return model
```

### Feature selection

#### No pre-selection post PCA

```python
train = df1.iloc[:632]
test = df1.iloc[632:]
```

```python
def evaluate_prophet_model(features, train_data):
    model = Prophet()
    for feature in features:
        model.add_regressor(feature)
    
    initial = 450  # This is an example, adjust based on your specific needs
    period = '63 days'  # half of horizon
    horizon = '126 days'  # Equivalent to proportion of test set
    
    model.fit(train_data)
    df_cv = cross_validation(model, initial=f'{initial} days', period=period, horizon=horizon, parallel="processes")
    df_p = performance_metrics(df_cv)
    
    return df_p['mae'].mean()
```

```python
predictors_df = train.drop(['ds', 'y'], axis=1)

candidate_predictors = predictors_df.columns.tolist()

selected_features = []
best_mae = float('inf')
best_features = []

# Forward feature selection loop
while True:
    improvement = False
    for feature in candidate_predictors:
        if feature not in selected_features:
            trial_features = selected_features + [feature]
            mae = evaluate_prophet_model(trial_features, train)
            
            if mae < best_mae:
                best_mae = mae
                best_features = trial_features
                improvement = True
    
    if improvement:
        selected_features = best_features
        print(f"New best model with features {best_features} with MAE: {best_mae}")
    else:
        break

print(f"Final best model includes features: {selected_features} with MAE: {best_mae}")
```
