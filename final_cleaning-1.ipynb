```python
import pandas as pd
df_1 = pd.read_csv('feature_selection_top_states.csv', index_col = 'Date', parse_dates = True).drop(columns=['Unnamed: 0'])
df_2 = pd.read_csv('multivariate_eng_df.csv', index_col = 'Date', parse_dates=['Date']).drop(columns=['Unnamed: 0'])
df_2.head()
```

```python
fl_df = df_1[df_1['PROVINCE_STATE_NAME'] == 'Florida']
ca_df = df_1[df_1['PROVINCE_STATE_NAME'] == 'California']
nc_df = df_1[df_1['PROVINCE_STATE_NAME'] == 'North Carolina']
pn_df = df_1[df_1['PROVINCE_STATE_NAME'] == 'Pennsylvania']
tx_df = df_1[df_1['PROVINCE_STATE_NAME'] == 'Texas']
```

```python
# Find indices of rows with negative PEOPLE_POSITIVE_NEW_CASES_COUNT
negative_indices = fl_df[fl_df['PEOPLE_POSITIVE_NEW_CASES_COUNT'] < 0].index

# Print rows surrounding negative values
for date in negative_indices:
    surrounding_dates = [date - pd.Timedelta(days=1), date, date + pd.Timedelta(days=1)]
    surrounding_rows = fl_df.loc[fl_df.index.intersection(surrounding_dates)]

x = pd.DataFrame(surrounding_rows)
x.head()
```

```python
def correct_negative_values(df):
    df['PEOPLE_POSITIVE_NEW_CASES_COUNT'] = df['PEOPLE_POSITIVE_NEW_CASES_COUNT'].apply(lambda x: x*-1 if x < 0 else x)
    return df
```

```python
fl_df_2 = correct_negative_values(fl_df)
fl_df_2.head()
```

```python
# Find indices of rows with negative PEOPLE_POSITIVE_NEW_CASES_COUNT
negative_indices = fl_df_2[fl_df_2['PEOPLE_POSITIVE_NEW_CASES_COUNT'] < 0].index

# Print rows surrounding negative values
for date in negative_indices:
    surrounding_dates = [date - pd.Timedelta(days=1), date, date + pd.Timedelta(days=1)]
    surrounding_rows = fl_df_2.loc[fl_df_2.index.intersection(surrounding_dates)]
    print(surrounding_rows)
```

```python
fsts_2 = correct_negative_values(df_1)
mved_2 = correct_negative_values(df_2)
```
